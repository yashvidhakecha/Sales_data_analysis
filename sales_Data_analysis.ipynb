```python
df = pd.read_excel("Sales Data_PDA_4052.xlsx", header=1)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['value_£'].describe()
```

```python
sales_by_person = df.groupby('sales_person')['value_£'].sum().sort_values(ascending=False)
sales_by_person
```

```python
plt.figure(figsize=(10,6))
sales_by_person.plot(kind='bar')
plt.title("Total Sales by Salesperson")
plt.xlabel("Sales Person")
plt.ylabel("Total Sales Value")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='priority', y='value_£', data=df)

plt.title("Sales Value by Priority")
plt.xlabel("Priority Level")
plt.ylabel("Sales Value")

plt.show()
```

```python
priority_counts = df['priority'].value_counts()
priority_counts
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='priority', data=df)

plt.title("Number of Orders by Priority")
plt.xlabel("Priority")
plt.ylabel("Count")

plt.show()
```

```python
sales_over_time = df.groupby('date')['value_£'].sum()

plt.figure(figsize=(10,5))
sales_over_time.plot()

plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")

plt.show()
```

```python
numeric_df = df[['value_£']]

corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")

plt.show()
```
